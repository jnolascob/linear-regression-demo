```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos el dataset
boston = load_boston()
print(boston.DESCR)
```

Fórmula para minimizar el error cuadrático medio (MCO)
$\beta = (X^{T}X)^{-1}X^{T}Y$

```python
tempX = boston.data[:, 5]
tempY = boston.target
# print(tempY)

X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

plt.scatter(X, Y, alpha=0.3)

plt.xlabel("# medio de habitaciones")
plt.ylabel("Valor medio in $1000")

# plt.show()

# # añadir columna de 1s para el término independiente
X = np.array([np.ones(506), X]).T
print(X)

B = np.linalg.inv(X.T @ X) @ X.T @ Y

# B

plt.plot([4, 9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c="blue")


plt.show()
```

```python
# Modelo encontrado
def getPriceFromRoom(number_rooms):
    return B[0] + B[1] * number_rooms    
```
